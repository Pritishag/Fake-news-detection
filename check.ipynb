```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
## panda for importing dataset 
## seaborn and matplotlib for data visualization 
```

```python
data = pd.read_csv('News.csv',index_col=0) ## importing downloaded data 
data.head() ## returns number of data from the top
```

```python
data.shape ##Returns tuple of shape (Rows, columns) of dataframe/series
```

```python
data = data.drop(["title", "subject","date"], axis = 1)
## since title , text and label wont be helpful in identification we rop them 
```

```python
data.isnull().sum()
## we check for null values 
```

```python
## shuffle the dataset to prevent the model to get bias and drop index as it is not useful 
data = data.sample(frac=1) 
data.reset_index(inplace=True) 
data.drop(["index"], axis=1, inplace=True)
```

```python
from tqdm import tqdm
import re
import nltk
nltk.download('punkt') 
nltk.download('stopwords') 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from wordcloud import WordCloud
```

```python
def preprocess_text(text_data): 
    preprocessed_text = [] 
    for sentence in tqdm(text_data):
        sentence = re.sub(r'[^\w\s]', '', sentence) 
        preprocessed_text.append(' '.join(token.lower()
                            for token in str(sentence).split()
                                  if token not in stopwords.words('english')))          
                                            
        
    return preprocessed_text
```

```python
preprocessed_review = preprocess_text(data['text'].values) 
data['text'] = preprocessed_review
```

```python
from sklearn.feature_extraction.text import CountVectorizer
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus) 
    bag_of_words = vec.transform(corpus) 
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
    for word, idx in vec.vocabulary_.items()] 
    words_freq = sorted(words_freq, key=lambda x: x[1], 
    reverse=True) 
    return words_freq[:n]
common_words = get_top_n_words(data['text'], 20) 
df1 = pd.DataFrame(common_words, columns=['Review', 'count']) 
df1.groupby('Review').sum()['count'].sort_values(ascending=False).plot(     
    kind='bar', 
    figsize=(10, 6), 
    xlabel="Top Words", 
    ylabel="Count", 
    title="Bar Chart of Top Words Frequency"
) 

 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
x_train, x_test, y_train, y_test = train_test_split(data['text'], 
 data['class'], 
 test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer() 
x_train = vectorization.fit_transform(x_train) 
x_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression() 
model.fit(x_train, y_train) 
# testing the model 
print(accuracy_score(y_train, model.predict(x_train))) 
print(accuracy_score(y_test, model.predict(x_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier() 
model.fit(x_train, y_train) 
# testing the model 
print(accuracy_score(y_train, model.predict(x_train))) 
print(accuracy_score(y_test, model.predict(x_test))) 
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier() 
model.fit(x_train, y_train) 
# testing the model 
print(accuracy_score(y_train, model.predict(x_train))) 
print(accuracy_score(y_test, model.predict(x_test))) 
```

```python
 # Confusion matrix of Results from Decision Tree classification 
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, model.predict(x_test)) 
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, 
 display_labels=[False, True]) 
cm_display.plot() 
plt.show()
```
